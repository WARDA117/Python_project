```python
# plotpie
df["score"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("score Distribution")
plt.ylabel("")  # removes y-label
plt.show()
```

```python
id = df["id"].value_counts().nlargest(5)   # show only top 5
id.plot.pie(autopct="%1.1f%%")
plt.title("Top 5 Categories")
plt.ylabel("")
plt.show()
```

```python
P_value = df3["P_value"].value_counts().nlargest(2)   # show only top 2
P_value.plot.pie(autopct="%1.1f%%")
plt.title("Top 2 Categories")
plt.ylabel("")
plt.show()
```

```python
#lineplot
plt.plot(df["score"], marker = "*", color = "red",linestyle = "--", linewidth = 2)
plt.grid()
plt.title("score Distribution")
plt.xlabel("Index")
plt.ylabel("score")
```

```python
#lineplot
plt.plot(df3["Connectivity"], marker = "*", color = "purple",linestyle = "--", linewidth = 2)
plt.grid()
plt.title("Connectivity Distribution")
plt.xlabel("Index")
plt.ylabel("Connectivity")
```

```python
#lineplot
plt.plot(df2["P_value"], marker = "*", color = "blue",linestyle = "--", linewidth = 1)
plt.grid()
plt.title("P_value Distribution")
plt.xlabel("Index")
plt.ylabel("P_vale")
```

```python

plt.plot(df["id"], df["score"], label = "GEN", marker = "o")
plt.plot(df3["Connectivity"], df2["P_value"], label = "relation",marker = "s" )
plt.xlabel("id")
plt.ylabel("score")
plt.title("Connectivity_P_value")
plt.legend()
plt.show()
```

```python
# 3d 

axs = plt.axes(projection = "3d")

axs.scatter(df3["Connectivity"], df3["P_value"])
axs.set_xlabel("Connectivity")
axs.set_ylabel("P_value")
axs.set_zlabel("Unnamed:0")
axs.set_title("3D Scatter plot")
plt.show()
```
