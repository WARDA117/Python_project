SEABORN

A Python data visualization library built on top of Matplotlib.

It provides a higher-level interface with simpler syntax to create statistically informative and aesthetically pleasing plots.

It integrates well with Pandas DataFrames, making it easier to visualize data directly from them.

It comes with built-in themes and color palettes, so plots look better with less effort compared to raw Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()
```

```python
# scatter

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size", palette= "Set1")
sns.set_style("dark")
sns.set_context("notebook")
```

```python
sns.scatterplot(data=tips, x="tip", y="size", hue="smoker", palette="Set2")
sns.set_style("dark")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="time", hue="sex", palette= "Set1", style="smoker")
```

```python
sns.scatterplot(data=tips, x="day", y="time", hue="smoker", palette= "Set1", style="total_bill")
```

```python
# stripplot

sns.stripplot(data = tips, x= "total_bill", y = "tip", hue = "smoker", dodge=True, jitter=True)
```

```python
# swarm plot

sns.set_style("whitegrid")
sns.set_context("talk")
sns.swarmplot(data=tips, x="smoker", y="day", hue="size", dodge=True)
sns.despine(left=True)
```

```python
# univariant

sns.histplot(data=tips, x= "total_bill", hue="size", multiple="stack", kde=True)
```

```python
#regression plot
sns.set_context("notebook")
sns.regplot(data = tips, x = "size", y = "size", color= "purple")
```

```python
sns.set_style("whitegrid")
sns.lineplot(data=tips, x="total_bill", y="size", style="day", hue = "time")
```

```python
# joint plot

sns.jointplot(data= tips, x = "total_bill", y="size", hue="time", kind="kde")
```

```python
# barplot # similar to histogram

sns.barplot(data=tips, x="time", y="total_bill", hue="size", estimator=np.sum)
```

```python
#countplot

sns.countplot(data=tips, x="day", hue="smoker", palette="Set1")
```

```python
# outliers detection # boxplot


sns.boxplot(data=tips, x="smoker", y="tip", hue = "time")
```

```python
# violin plot # width shows the density

sns.violinplot(data= tips, x="sex", y="size", split=True, hue="time")
```

```python
#swarm plot

sns.swarmplot(data=tips, x="total_bill", y="time", hue="smoker", dodge=True)
```

```python
# kdeplot
sns.kdeplot(data=tips, x="total_bill")
sns.despine()
```

```python
#2d plot
column = ["total_bill","tip","size"]

tips[column].corr()
```

```python
# heatmap
sns.heatmap(tips[column].corr(), annot=True, cmap="PRGn", center=0, linewidths="2",vmin=-1, vmax=1)
```

```python
# regression 

sns.regplot(data=tips, x="total_bill", y="tip")
```

```python
# rug plot

sns.rugplot(data=tips, x="total_bill", height=3)
```

```python
# pair plot

sns.pairplot(tips, hue="size", diag_kind="kde", markers=["o", "s", "d"])
```

```python
# pair grid

graph = sns.PairGrid(data =tips, hue="sex", palette="Set1")
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
graph.map_upper(sns.scatterplot)
```
